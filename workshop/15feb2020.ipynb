# generators

```python
## functions
## instead of return using yield keyword
def generator():
    yield 1
    yield 2
    yield 3
for value in generator ():
    print(value)
## generator()
```

```python
def  gen():
    s=2
    while (1==1):## 1==1 or True placed 
        s**=2
        yield s
a=gen()
for i in range(5):
    print(next(a))
        
```

```python
def gen():
    s=["s","a","i"]
    while True:
        yield s
a=gen()
for i in range(5):
    print(next(a))
```

# filters

```python
## it is used to separate what we u want 
## it takes two arguments 
## first function name 
## iterables as a tuples and sets and lists 
## syntax is filter(function,iterable)
```

```python
def even(n):
    if(n%2==0):
        return n+n
print(list(filter(even,[1,2,3,4,5])))
```

```python
def prime(n):
    count=0
    for i in range(1,n//2+1):
        if(n%i==0):
            count=count+1
    if(count==1):
        return n
print(list(filter(prime,range(1,50))))
```
